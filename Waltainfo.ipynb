```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.waltainfo.com/?locale=am"

headers = {
    "User-Agent": "Mozilla/5.0"
}


response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

links = []

for a in soup.find_all("a", href=True):

    title = a.get_text(strip=True)
    url = a["href"]

    if title and len(title) > 25 and "waltainfo.com" in url:

        if url.startswith("/"):
            url = "https://www.waltainfo.com" + url

        links.append({
            "title": title,
            "url": url
        })


unique_links = {x["url"]: x for x in links}.values()

articles = []

for item in unique_links:

    try:
        r = requests.get(item["url"], headers=headers)
        s = BeautifulSoup(r.text, "html.parser")

       
        articles.append({
            "title": item["title"],
            "url": item["url"],
        })

    except Exception as e:
        print("Error:", e)

df = pd.DataFrame(articles)
df.to_csv("walta_level1.csv", index=False, encoding="utf-8")

print(df.head())
```
