```python
df = pd.DataFrame({
    "name": ["sihu", "hayu", "SIHU", "ekru", "hayu", None],
    "age": [20, np.nan, 20, 30, np.nan, 25],
    "city": ["addis ababa", "ADDIS ABABA", "Addis Ababa", "bahir dar", "ADDIS ABABA", "bahir dar"]
})
df["age"] = df["age"].fillna(df["age"].median())
df["name"] = df["name"].fillna("unknown")
df["name"] = df["name"].str.lower().str.strip()
df["city"] = df["city"].str.title().str.strip()
df = df.drop_duplicates()
print(df)
```

```python
df = pd.DataFrame({
    "sales": [100, 120, 130, 150, 160, 170, 180, 200, 1000]
})
Q1 = df["sales"].quantile(0.25)
Q3 = df["sales"].quantile(0.75)

print("Q1:", Q1)
print("Q3:", Q3)

IQR = Q3 - Q1
print("IQR:", IQR)

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

outliers = df[(df["sales"] < lower_bound) | (df["sales"] > upper_bound)]

print( outliers)
```

```python
df = pd.DataFrame({
    "name": ["sihu", "HAYU", "sihu", "ekru", None, "hayu"],
    "age": [20, np.nan, 20, 30, 200, 25],
    "sales": [100, 120, 100, 150, 5000, np.nan],
    "city": ["addis ababa", "ADDIS ABABA", "Addis Ababa", "bahir dar", "ADDIS ABABA", "bahir dar"]
})
print( df.isnull().sum())
df["age"] = df["age"].fillna(df["age"].median())
df["sales"] = df["sales"].fillna(df["sales"].median())
df["name"] = df["name"].fillna("unknown")
df = df.drop_duplicates()
df["name"] = df["name"].str.lower().str.strip()
df["city"] = df["city"].str.title().str.strip()
Q1 = df["sales"].quantile(0.25)
Q3 = df["sales"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower:", lower_bound)
print("Upper:", upper_bound)

df = df[(df["sales"] >= lower_bound) & (df["sales"] <= upper_bound)]
print(df)
```
