```python
import pandas as pd
```

```python
df = pd.DataFrame({"student_id":[101,102,103,102,104,101],
	  "name":["Abel","Sara","Jhon","Sara","Helen","Abel"],
	  "score":[80,90,75,90,88,80]
	})
```

```python
print(df.duplicated())
```

```python
print("Duplicate Count:",df.duplicated().sum())
```

```python
df_clean = df.drop_duplicates()
print(df_clean)
```

```python
df_last = df.drop_duplicates(keep='last')
print(df_last)
```

```python
df = pd.DataFrame({"id":[1,2,3,4],
      "email":["a@gmail.com","b@gmail.com","a@gmail.com","c@gmail.com"]             })
```

```python
duplicates =df[df.duplicated(subset=["email"])]
print(duplicates)
```

```python
clean_df = df.drop_duplicates(subset =["email"])
print(clean_df)
```

```python
df = pd.DataFrame({"city":["addis abeba","ADDIS ABEBA","Addis Abeba","adDia abAba"]})
df["upper"]= df["city"].str.upper()
print(df)
```

```python
df["lower"] =df["city"].str.lower() 
df["title"] = df["city"].str.title()
print(df)
```

```python
df = pd.DataFrame({"name":["Abel","Sara "," John "," Helen"]})
df["name"]= df["name"].str.strip()
print(df)
```

```python
df = pd.DataFrame({"date":["2026-01-05","05/02/2026","March 10,2026","2026.04.15"]})
df["date"] = pd.to_datetime(df["date"], errors="coerce")
df["formated"] = df["date"].dt.strftime("%Y-%m-%d")
df["year"] = df["date"].dt.year
df["month"]= df["date"].dt.month
print(df)
```

```python
df = pd.DataFrame({"weight_kg":[55,70,85,100]})
df["weight_lb"] = (df["weight_kg"]*2.20462).round(2)
print(df)
```

```python
df = pd.DataFrame({"temp_c":[0,25,37,100]})
df["temp_f"] = (df["temp_c"]*9/5)+32
print(df)
```

```python
import numpy as np

df = pd.DataFrame({"name":["Abel","Sara","None","Jhon"],
                   "age":[20,np.nan,25,30]})
print(df.isnull())
print()
print(df.isnull().sum())
print()
df["age"]= df["age"].fillna(df["age"].mean())
print()
df["name"]= df["name"].fillna("unknown")
print(df)
```

```python
df = pd.DataFrame({"gender":["Male","male","MALE","Female","female"]})
df["gender"]= df["gender"].str.title()
print(df)
```

```python
df = pd.DataFrame({
    "salary":[3500,4000,4200,3900,4100,45000]
})
Q1 = df["salary"].quantile(0.25)
Q3 = df["salary"].quantile(0.75)
IQR = Q3-Q1

lower = Q1-1.5 *IQR
upper =Q3+1.5 *IQR

outliers = df[(df["salary"] < lower)| (df["salary"]) > upper]
print(outliers)

clean_df = df[(df["salary"] >= lower) & (df["salary"] >= lower)]
print(clean_df)
```
