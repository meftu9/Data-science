```python
import pandas as pd
import numpy as np
```

```python
data = {
'Patient_ID':['P001','P002','P003','P004','P004'],
'Full_Name':['Abebe Kebede',' Hana Tesfaye ','Samuel Bekele','Meron Alemu','Meron Alemu'],

'Age':[35,np.nan,42,29,29],
'Address':[
'Bole Road, Addis Ababa, Bole, 03, H125',
'CMC Road, Addis Ababa, Yeka, 10, H201',
'Piassa, Addis Ababa, Arada, 01, H505',
'Megenagna, Addis Ababa, Bole, 05, H901',
'Megenagna, Addis Ababa, Bole, 05, H901'
]
}
df = pd.DataFrame(data)
print(df)
```

```python
print(df.isnull().sum())
df["Age"] = df["Age"].fillna(df["Age"].mean())
print(df.drop_duplicates())
df["Full_Name"] = df["Full_Name"].str.strip()
df["First_Name"] = df["Full_Name"].str.split().str[0]
df["Last_Name"] = df["Full_Name"].str.split().str[-1]

df["Address"] = df["Address"].str.strip()

df[["Street", "City", "Subcity", "Woreda", "House_No"]] = df["Address"].str.split(",", expand=True)

print(df)
```

```python
data = {
'Month':['Jan','Jan','Feb','Feb','Mar','Mar'],
'Product':['Laptop','Phone','Laptop','Phone','Laptop','Phone'],
'Sales':[15,np.nan,18,20,15,25]
}

df = pd.DataFrame(data)
print(df)
```

```python

pivoted = pd.pivot(
    df,
    index="Month",
    columns="Product",
    values="Sales", 
)

pivoted = pivoted.fillna(0)
print(pivoted)
```

```python
data = {
'Employee_ID':['E1','E2','E3','E4'],
'Email':[
'abebe@gmail.com',
'hana@yahoo.com',
'samuel@outlook.com',
'meron@gmail.com'
]
}

df = pd.DataFrame(data)

df[["Username","Domain"]] = df["Email"].str.split("@", expand=True)
df["Company_Email"] = df["Username"] 
end_gmail = df[df["Email"].str.endswith("@gmail.com")]
print(end_gmail)

print()
print(df)
```

```python
data = {
'Region':['Addis','Addis','Adama','Adama','Hawassa','Bahir Dar'],
'Loan_Type':['Personal','Mortgage',
'Personal','Mortgage',
'Personal','Mortgage'],
'Status':['Approved','Rejected',
'Approved','Approved',
'Rejected','Approved']

}

df = pd.DataFrame(data)

pivoted = pd.pivot_table(
    df,
    index="Region",
    columns=["Loan_Type", "Status"],
    aggfunc="size",
    fill_value=0
)

print(pivoted)

approved_loans = df[df["Status"] == "Approved"].groupby("Region").size()
print()
print(approved_loans.idxmax())

```
